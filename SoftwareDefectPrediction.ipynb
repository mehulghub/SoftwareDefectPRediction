```python
df['defects']= np.where(df['defects']== True,1,0)
df
```

```python
### Train test split to avoid overfitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['defects'], axis=1),
    df['defects'],
    test_size=0.3,
    random_state=0)
```

```python
from sklearn.feature_selection import mutual_info_classif
# determine the mutual information
mutual_info = mutual_info_classif(X_train, y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
#let's plot the ordered mutual_info values per feature
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
from sklearn.feature_selection import SelectKBest
#No we Will select the  top 5 important features
sel_five_cols = SelectKBest(mutual_info_classif, k=5)
sel_five_cols.fit(X_train, y_train)
X_train.columns[sel_five_cols.get_support()]
```

```python
df_feat=pd.DataFrame(scaled_features, columns=X_train.columns)
df_feat.head()
```

KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier(n_neighbors=5)
```

```python
X_test= X_test[['l', 'd', 'e', 't', 'uniq_Opnd']]
```

```python
model1.fit(df_feat, y_train)

y_pred = model1.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
#Summary of the predictions made by the classifier
print("K-Nearest Neighbors Algorithm")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
#Accuracy score
from sklearn.metrics import accuracy_score
print("ACC: ",accuracy_score(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
knn_pipe = Pipeline([('mms', MinMaxScaler()),
                     ('knn', KNeighborsClassifier())])
params = [{'knn__n_neighbors': [3, 5, 7, 9],
         'knn__weights': ['uniform', 'distance'],
         'knn__leaf_size': [15, 20]}]
gs_knn = GridSearchCV(knn_pipe,
                      param_grid=params,
                      scoring='accuracy',
                      cv=5)
gs_knn.fit(df_feat, y_train)
gs_knn.best_params_
accuracy = gs_knn.best_score_
accuracy
```

```python
decisiontree_params = [{'max_depth': list(range(10, 15)), 'max_features': list(range(0,14))}]

clf = GridSearchCV(model2, decisiontree_params, cv = 10, scoring='accuracy')

clf.fit(df_feat, y_train)

clf.best_params_

accuracy = clf.best_score_
accuracy
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=100)
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)

model3.fit(df_feat, y_train)

y_pred = model3.predict(X_test)

#Summary of the predictions made by the classifier
print("Random Forests Algorithm")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
#Accuracy score
from sklearn.metrics import accuracy_score
print("ACC: ",accuracy_score(y_pred,y_test))
```

```python
rfc_params = [{'max_depth': list(range(10, 15)), 'max_features': list(range(0,14))}]

clf2 = GridSearchCV(model3, rfc_params, cv = 10, scoring='accuracy')

clf2.fit(df_feat, y_train)

clf2.best_params_

accuracy = clf2.best_score_
accuracy
```

```python
from sklearn.linear_model import LogisticRegression
model4 = LogisticRegression()
model4.fit(df_feat, y_train)
y_pred = model4.predict(X_test)

print("Logistic Regression Algorithm")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("ACC: ",accuracy_score(y_pred,y_test))


# Support Vector Machine (SVM)
from sklearn.svm import SVC
model5 = SVC()
model5.fit(df_feat, y_train)
y_pred = model5.predict(X_test)

print("Support Vector Machine Algorithm")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("ACC: ",accuracy_score(y_pred,y_test))


# Naive Bayes
from sklearn.naive_bayes import GaussianNB
model6 = GaussianNB()
model6.fit(df_feat, y_train)
y_pred = model6.predict(X_test)

print("Naive Bayes Algorithm")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("ACC: ",accuracy_score(y_pred,y_test))
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you have your models defined as model1, model2, model3, etc.
# and your data as X and y (features and target variable)

# Define the cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Iterate through the models
models = [model1, model2, model3, model4, model5, model6]  # Replace with your model variables
model_names = ['KNN', 'Decision Tree', 'Random Forest', 'Logistic Regression', 'SVM', 'Naive Bayes']

for model, model_name in zip(models, model_names):
    # Cross-validation for accuracy
    accuracy_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    print(f"{model_name} - Accuracy (Cross-Validation): {accuracy_scores.mean():.4f} (+/- {accuracy_scores.std() * 2:.4f})")

    # Calculate other metrics using cross_val_predict
    from sklearn.model_selection import cross_val_predict
    y_pred = cross_val_predict(model, X_train, y_train, cv=cv)

    precision = precision_score(y_train, y_pred)
    recall = recall_score(y_train, y_pred)
    f1 = f1_score(y_train, y_pred)

    print(f"{model_name} - Precision: {precision:.4f}")
    print(f"{model_name} - Recall: {recall:.4f}")
    print(f"{model_name} - F1-Score: {f1:.4f}")
    print("-" * 30)
```
